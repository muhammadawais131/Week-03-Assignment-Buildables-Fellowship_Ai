### CODE FOR CHATBOT :

```python
# üåø Eco-Guide: Sustainable Living Assistant Chatbot

# Install required library
!pip install groq

# Import necessary modules
from groq import Groq
import getpass
import sys

# --- CONFIGURATION ---
# Initialize the Groq client with your API Key
print("Please enter your Groq API key to continue. (Your typing will be hidden)")
api_key = getpass.getpass('Groq API key...: ').strip()

if not api_key:
    print("Error: No API key provided. Exiting.")
    sys.exit(1)

try:
    client = Groq(api_key=api_key)
    # Test the connection with a simple request
    test_completion = client.chat.completions.create(
        messages=[{"role": "user", "content": "Hello"}],
        model="llama-3.1-8b-instant",  # Using a currently available model
        max_tokens=5,
    )
    print("API connection successful!")
except Exception as e:
    print(f"Failed to initialize Groq client: {e}")
    print("Please check your API key and try again.")
    sys.exit(1)

# Define the AI model - using a currently available model
MODEL_NAME = "llama-3.1-8b-instant"  # Updated to a currently supported model

# Alternative models you could try:
# - "llama-3.1-70b-versatile"
# - "llama-3.1-8b-instant"
# - "mixtral-8x7b-32768" (if it becomes available again)
# - "gemma2-9b-it"

SYSTEM_PROMPT = """
You are 'Eco-Guide', a friendly and knowledgeable Sustainable Living Assistant.
Your goal is to provide practical, actionable, and non-judgmental advice to help users live more eco-friendly lives.
You specialize in:
- Waste reduction (food, plastic, etc.)
- Recycling and composting guidelines
- Sustainable product alternatives and DIY solutions
- General education on environmental impact

Always be encouraging and focus on progress, not perfection. If a user asks about something outside your scope (like math or movies), politely decline and steer the conversation back to sustainability.
If a question is complex, break it down into clear, step-by-step instructions.
"""

# --- MAIN CHAT FUNCTION ---
def run_eco_guide():
    messages = [
        {"role": "system", "content": SYSTEM_PROMPT},
    ]

    print("\n" + "="*50)
    print("üåø Welcome to Eco-Guide! Your Sustainable Living Assistant.")
    print("Let's talk about recycling, reducing waste, and eco-friendly swaps!")
    print("Type 'quit' at any time to end the conversation.")
    print("="*50 + "\n")

    while True:
        user_input = input("You: ").strip()

        if not user_input:
            continue

        if user_input.lower() == 'quit':
            print("\nEco-Guide: Thanks for chatting! Remember, every small change helps. üåé")
            break

        messages.append({"role": "user", "content": user_input})

        try:
            chat_completion = client.chat.completions.create(
                messages=messages,
                model=MODEL_NAME,
                temperature=0.5,
                max_tokens=1024,
            )

            ai_response = chat_completion.choices[0].message.content
            print(f"\nEco-Guide: {ai_response}\n")
            messages.append({"role": "assistant", "content": ai_response})

        except Exception as e:
            print(f"\nSorry, I encountered an error: {e}")
            print("This might be due to an invalid API key or model availability issues.")
            break


run_eco_guide()
```

### 5. Prompting Technique
I will use a combination of:

System Prompt Persona: To firmly establish the bot's role, knowledge boundaries, and tone (friendly, encouraging, and practical).

Few-shot Prompting: To prime the model with examples of high-quality, actionable responses for common queries.

Chain-of-thought prompting: For questions that require a multi-step explanation (e.g., "How do I start a compost bin?").
